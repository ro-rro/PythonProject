## Exercise 2: 

Create bins for each full number of earthquake magnitude (for instance, the first bin is (0,1], the second is (1,2], and so on) with the ml magnitude type and count how many are in each bin.

First step for this exercise is to create groups for the intervals. The intervals will be 0-1, 1-2, 2-3, etc. until the max ml is reached. Bin_count is used to count the number of values that correspond to each bin. 

```python
ml_quakes = earthquake[earthquake["magType"] == "ml"] # filter for only ml values

bins = np.arange(0, ml_quakes["mag"].max() + 1, 1)

# groups the values into intervals 
mag_bins = pd.cut(ml_quakes["mag"], bins=bins) # cut is used to assign each earthquake to a magnitude range

# counts the number of earthquakses that are in each interval. 
bin_count = mag_bins.value_counts().sort_index()

bin_count
```

The max ml value was 6, 6 different bins were created. The bin with the most was 1-2, the one with the least was 5-6. If a histogram were created for this problem, the data would skewed be skewed to the right. 

## Exercise 4: 

Build a crosstab with the earthquake data between the tsunami column and the magType column. Rather than showing the frequency count, show the maximum magnitude that was observed for each combination. Put the magnitude type along the columns.

A crosstab is created to examine the relationship between tsunami and magnitude types (earthquakes). Instead of frequency, max value for each magnitude type was found (aggregation max)

```python
cross_tab = pd.crosstab(
    earthquake["tsunami"], 
    earthquake["magType"], 
    values = earthquake["mag"], 
    aggfunc = "max")

cross_tab
```

Each cell contains the largest earthquake magnitude which is done using the values and through aggregation. The magnitude type is shown as the columns and the comparison can be seen if there was or wasn't a tsunami (0 or 1). Larger magnitude is usually associated with a tsumani, but this is only one factor that is involved.

## Exercise 7: 

Calculate the z-scores for each numeric column of Amazon’s data (ticker is AMZN) in Q4 2018 using apply().

Question is only asking for dates in Q4, those would be between November 1st to the end of December. Filter can be created to only see Amazon data during those dates. Since z-score is being used, numeric data from the data set is only going to be used, date and ticker would no longer be relavent. Z-score can then be applied to all of numeric values, using equation. 

```python
faang["date"] = pd.to_datetime(faang["date"])

amzn_q4 = faang[
    (faang["ticker"] == "AMZN") &
    (faang["date"] >= "2018-10-01") &
    (faang["date"] <= "2018-12-31")]

# numeric values are the only ones being selected/relavent
num_col = amzn_q4.select_dtypes(include = np.number) 

z_score = num_col.apply(
    lambda x: (x - x.mean())/ x.std()) # z-score equation

z_score
```

The following opens the covid_19 data. 

```python
covid_data = pd.read_csv("/Users/rosienavarro/Data_Science/Hands-On-Data-Analysis-with-Pandas-2nd-edition-master/ch_04/exercises/covid19_cases.csv")

covid_data
```

Parsed dataRep converted to datetime. Datetime convers the text into date values. 

```python
covid_data["date"] = pd.to_datetime(
    covid_data["dateRep"], 
    format = "%d/%m/%Y")

# sets date as index
covid_data = covid_data.set_index("date")
```

The following converts/replaces United states and United Kingdom to their abbreviations. 

```python
covid_data["countriesAndTerritories"] = covid_data[
    "countriesAndTerritories"].replace({
        "United_States_of_America" : "USA", 
        "United_Kingdom" : "UK"})
```

The following sorts the index. To make sure the following steps worked, chart is displayed again. 

```python
covid_data = covid_data.sort_index()
covid_data
```

### For the five countries with the most cases (cumulative), find the day with the largest number of cases.

There are two things done for this problem, country and cases are grouped. Then the sum of the top five are displayed. Look for the countries with the most cases. 

```python
top_5 = covid_data\
    .groupby('countriesAndTerritories').cases.sum()\
    .nlargest(5).index

top_5

covid_data[covid_data.countriesAndTerritories.isin(top_5)]\
    .groupby('countriesAndTerritories').cases.idxmax()
```

The following shows the contries with the most cumulative cases and the dates they had the most cases. 


### Find the 7-day average change in COVID-19 cases for the last week in the data for the five countries with the most cases.

The final 7 days are slected for the top 5 countries with the most cases. 

```python
top_5_data = covid_data[
    covid_data["countriesAndTerritories"].isin(top_5)]

average_day = (
    top_5_data.groupby("countriesAndTerritories")["cases"]
    .rolling(7)
    .mean()
    .reset_index())

last_week = average_day.groupby(
    "countriesAndTerritories").tail(7)

last_week
```

### Find the first date that each country other than China had cases.

Goal of this problem is to find the date when the first case appear in each country. This can be done by grouping countries with cases that are more than 0. 

```python
first_case = (
    covid_data[
        (covid_data["cases"] > 0) &
        (covid_data["countriesAndTerritories"] != "China")]
    .groupby("countriesAndTerritories")
    .apply(lambda x: x.index.min()))

first_case
```

### Rank the countries by cumulative cases using percentiles.
Countries with values closer to 1 have higher cumulative case counts compared to the other countries. 

```python
country_total = (
    covid_data.groupby("countriesAndTerritories")["cases"]
    .sum())

percentile = country_total.rank(pct = True) # pct converts the rank between 0-1. 
percentile.sort_values(ascending = False)
```

The US has the highest percentile and Anguilla has the lowest percentile of covid cases (0.0047).
